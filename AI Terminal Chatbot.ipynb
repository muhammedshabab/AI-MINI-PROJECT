```python
from openai import OpenAI

client = OpenAI(
    base_url="http://localhost:11434/v1",
    api_key="ollama"
)
while True:
    user=input("enter something")
    if user.lower()=="exit":
        break
    
    response = client.chat.completions.create(
    model="llama3:8b",
    messages=[{"role": "user", "content": user}]
)
    print(response.choices[0].message.content)
```
