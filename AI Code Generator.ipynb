```python
from openai import OpenAI

client = OpenAI(
    base_url="http://localhost:11434/v1",
    api_key="ollama"
)

print("AI Code Generator (type 'exit' to quit)\n")

while True:
    user = input("Enter your coding request: ")

    if user.lower() == "exit":
        break

    response = client.chat.completions.create(
        model="llama3:8b",
        messages=[
            {
                "role": "system",
                "content": "You are an AI Code Generator. Generate clean, correct programming code based on the user's request. Explain shortly if needed."
            },
            {
                "role": "user",
                "content": user
            }
        ]
    )

    print("\nGenerated Code:\n")
    print(response.choices[0].message.content)
    print("\n" + "-"*50 + "\n")
```
