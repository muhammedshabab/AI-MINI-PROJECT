```python
from openai import OpenAI

client = OpenAI(
    base_url="http://localhost:11434/v1",
    api_key="ollama"
)

print("AI Email Writer (type 'exit' to quit)\n")

while True:
    user = input("Enter email topic or request: ")

    if user.lower() == "exit":
        break

    response = client.chat.completions.create(
        model="llama3:8b",
        messages=[
            {
                "role": "system",
                "content": "You are an AI Email Writer. Write clear, professional, and well-structured emails based on the user's request."
            },
            {
                "role": "user",
                "content": user
            }
        ]
    )

    print("\nGenerated Email:\n")
    print(response.choices[0].message.content)
    print("\n" + "-"*50 + "\n")
```
