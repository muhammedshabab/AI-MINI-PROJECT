```python
from openai import OpenAI

client = OpenAI(
    base_url="http://localhost:11434/v1",
    api_key="ollama"
)

print("AI Text Summarizer (type 'exit' to quit)\n")

while True:
    user = input("Enter text to summarize: ")

    if user.lower() == "exit":
        break

    response = client.chat.completions.create(
        model="llama3:8b",
        messages=[
            {
                "role": "system",
                "content": "You are an AI Text Summarizer. Summarize the user's text clearly and briefly."
            },
            {
                "role": "user",
                "content": user
            }
        ]
    )

    print("\nSummary:\n")
    print(response.choices[0].message.content)
    print("\n" + "-"*50 + "\n")
```
