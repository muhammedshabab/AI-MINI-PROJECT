```python
from openai import OpenAI

client = OpenAI(
    base_url="http://localhost:11434/v1",
    api_key="ollama"
)

print("AI Grammar Corrector (type 'exit' to quit)\n")

while True:
    user = input("Enter text to correct: ")

    if user.lower() == "exit":
        break

    response = client.chat.completions.create(
        model="llama3:8b",
        messages=[
            {
                "role": "system",
                "content": "You are an AI Grammar Corrector. Correct the grammar, spelling, and sentence structure of the user's text and return the corrected version."
            },
            {
                "role": "user",
                "content": user
            }
        ]
    )

    print("\nCorrected Text:\n")
    print(response.choices[0].message.content)
    print("\n" + "-"*50 + "\n")
```
