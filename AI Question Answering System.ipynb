```python
from openai import OpenAI

client = OpenAI(
    base_url="http://localhost:11434/v1",
    api_key="ollama"
)

print("AI Tutor - Question Answering System (type 'exit' to quit)\n")

while True:
    user = input("Ask your question: ")

    if user.lower() == "exit":
        print("AI Tutor closed.")
        break

    response = client.chat.completions.create(
        model="llama3:8b",
        messages=[
            {
                "role": "system",
                "content": "You are an AI Tutor. Answer the user's questions clearly, step-by-step, in a simple way so students can easily understand."
            },
            {
                "role": "user",
                "content": user
            }
        ]
    )

    print("\nAnswer:\n")
    print(response.choices[0].message.content)
    print("\n" + "-"*50 + "\n")
```
